```python
def zeroshot_classifier(model, classnames, templates):
    with torch.no_grad():
        zeroshot_weights = []
        for classname in tqdm(classnames):
            texts = [template.format(classname) for template in templates] #format with class
            texts = clip.tokenize(texts).cuda() #tokenize
            class_embeddings = model.encode_text(texts) #embed with text encoder
            class_embeddings /= class_embeddings.norm(dim=-1, keepdim=True)
            class_embedding = class_embeddings.mean(dim=0)
            class_embedding /= class_embedding.norm()
            zeroshot_weights.append(class_embedding)
        zeroshot_weights = torch.stack(zeroshot_weights, dim=1).cuda()
    return zeroshot_weights
```

```python
def collate(x):
    imgs, labels = zip(*x)
    labels = torch.LongTensor(labels)
    return imgs, labels

val_images = datasets.ImageNet("~/ILSVRC2012/", split='val',)
single_loader = torch.utils.data.DataLoader(val_images, batch_size=1, shuffle=False, num_workers=0, collate_fn=collate)
val_loader = torch.utils.data.DataLoader(val_images, batch_size=64, shuffle=True, num_workers=0, collate_fn=collate)
```

```python
class ImageNetClassificationCLIP(torch.nn.Module):
    def __init__(self, model_name, device):
        super(ImageNetClassificationCLIP, self).__init__()
        self.device = device
        self.clip_model, self.transform = clip.load(model_name, device)
        try:
            with open('./zeroshot_imagenet_weights_'+model_name.replace('/','_')+'.pkl', 'rb') as fp:
                text_weights = pickle.load(fp)
        except Exception:
            text_weights = zeroshot_classifier(self.clip_model, imagenet_classes, imagenet_templates)
            with open('./zeroshot_imagenet_weights_'+model_name.replace('/','_')+'.pkl', 'wb+') as fp:
                pickle.dump(text_weights, fp)
        self.text_weights = torch.nn.Parameter(data=text_weights.to(torch.float32), requires_grad=True)
#         self.register_buffer('text_weights', text_weights.to(torch.float32)) 
#         print(self.text_weights.device)
        self.preprocess = T.Compose(self.transform.transforms[:-1])
        self.normalize = self.transform.transforms[-1]
#         self.postprocess = torch.nn.Sequential(torch.nn.Linear(512, 700),
#                                                torch.nn.ReLU(),
#                                                torch.nn.Linear(700, 900),
#                                                torch.nn.ReLU(),
#                                                torch.nn.Linear(900, 1000))
#         self.freeze()
    
    def freeze(self, fb=False):
        for param in self.clip_model.parameters():
            param.requires_grad = fb
        
#     def train(self):
#         super(ImageNetClassificationCLIP, self).train()
#         self.freeze(True)
    
    def forward(self, imgs):
        imgs = imgs.to(self.device)
        image_features = self.clip_model.encode_image(imgs).to(torch.float32)
#         print(image_features.shape, image_features.dtype)
        image_features = image_features/image_features.norm(dim=-1, keepdim=True)
#         logits = self.postprocess(image_features)
        logits = image_features @ self.text_weights
#         logits = self.pt*logits + (1-self.pt)*self.postprocess(image_features)
        return logits
    
vit_clip_model = ImageNetClassificationCLIP("ViT-B/32", device )
```

```python
class ImageNetClassificationTimm(torch.nn.Module):
    def __init__(self, model_name, device):
        super(ImageNetClassificationTimm, self).__init__()
        self.device = device
        self.model = timm.create_model(model_name, pretrained=True).to(device)
        self.model.eval()
        self.transform = create_transform(**resolve_data_config({}, model=self.model))
        self.preprocess = T.Compose(self.transform.transforms[:-1])
        self.normalize = self.transform.transforms[-1]
#         self.freeze()
        self.to(self.device)
        
    def freeze(self, fb=False):
        for param in self.model.parameters():
            param.requires_grad = fb
    
#     def train(self, ):
#         super(ImageNetClassificationTimm, self).train()
#         self.freeze(True)
    
    def forward(self, imgs):
        imgs = imgs.to(self.device)
        logits = self.model(imgs)
        return logits
    
vit_model = ImageNetClassificationTimm('vit_base_patch32_224', device)
```

```python
class MyDataParallel(torch.nn.DataParallel):
    def __getattr__(self, name):
        try:
            return super().__getattr__(name)
        except AttributeError:
            return getattr(self.module, name)

def finetune(model, train_loader, val_loader, device, lr=1e-3, eps = 0.01, epochs=2, adv=False):
    loss_fn = torch.nn.CrossEntropyLoss()
    model.freeze(False)
    model = MyDataParallel(model)
    model.to(device)
    optimizer = torch.optim.SGD(model.parameters(), lr=lr)
    test_acc = 0
    misses = 0
    min_lr = 1e-7
    check_interval = 100
    min_test_size = 15
    for e in range(epochs):
        total_loss = 0
        if e == 1:
            model.freeze(True)
        for i, (imgs, labels) in enumerate(tqdm(train_loader)):
            if i%check_interval == 0 and i > 0:
                
                hits, total = 0, 0
                for j, (test_imgs, test_labels) in enumerate(val_loader):
                    test_imgs, test_labels = test_imgs.to(device), test_labels.to(device)
                    if adv:
                        atk = torchattacks.PGD(model, eps=eps, alpha=eps/10, steps=15)
                        test_adv_imgs = atk(test_imgs, test_labels)
                        test_imgs = test_adv_imgs
                    preds = model(model.normalize(test_imgs))
                    a = hits/(total + 1e-4)
                    hits += torch.sum(preds.argmax(dim=-1) == test_labels)
                    total += len(labels)
                    if j>min_test_size and torch.abs(hits/total - a) < 1e-3:
                        break
                acc = hits/total
                print(f"Test accuracy at batch num {i} is {acc}")
                 
                if acc < test_acc:
                    if misses == 5:
                        lr = lr/2
                        if lr < min_lr:
                            return model
                        else:
                            optimizer = torch.optim.SGD(model.parameters(), lr=lr)
                            misses = 0
                    else:
                        misses += 1
                else:
                    test_acc = acc
                    misses = 0
                    
            optimizer.zero_grad()
            if adv:
                atk = torchattacks.PGD(model, eps=eps, alpha=eps/10, steps=15)
                adv_imgs = atk(imgs, labels)
                imgs = torch.cat([adv_imgs], dim=0)
                labels = torch.cat([labels], dim=0)
            logits = model(model.normalize(imgs))
            loss = loss_fn(logits, labels.to(model.device))
            total_loss += loss.detach()
            loss.backward()
            optimizer.step()
            if i % 20 == 0:
                print(f"Training loss at batch num {i} is {total_loss/(i+1)}")
            
        total_loss = total_loss/(i+1)
        print(f"Total loss is {total_loss}")
    return model
```
